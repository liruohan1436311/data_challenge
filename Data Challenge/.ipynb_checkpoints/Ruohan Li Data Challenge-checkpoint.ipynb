### 2. What is the average lease-up time for these two markets?

```python
# Load Data
msa1_occ = pd.read_excel("./raw_data/leaseup/MSA1.xlsx",sheet_name="Occ & Concession", header = 4, na_values = " ")
msa2_occ = pd.read_excel("./raw_data/leaseup/MSA2.xlsx",sheet_name="Occ & Concession", header = 4, na_values = " ")
```

```python
print(msa1_occ.shape)
print(msa2_occ.shape)
```

<p> Lease-up time is defined as the number of months since the property delivered to market to the first
month when the property reaches 90% occupancy.
<p> After looking through the dataset, for both markets, different worksheets share the same primary key, which is the row index. Therefore, we can aggregate the information across different worksheet through keeping track of the row index.We can get the delivered date from the "Property Status" worksheet and go through the "Occ and Concession" worksheet to count the number of months until the property reaches 90% occupancy.

```python
start_status = msa1_status.columns.get_loc("Apr-08")
end_status = msa1_status.columns.get_loc("Sep-19")
start_occ = msa1_occ.columns.get_loc("Apr-08")
end_occ = msa1_occ.columns.get_loc("Sep-19")
msa1_deliver_occ = msa1_occ[(((msa1_status == "LU").sum(axis = 1)) > 0) | (((msa1_status == "UC/LU").sum(axis = 1)) > 0)]
msa2_deliver_occ = msa2_occ[(((msa2_status == "LU").sum(axis = 1)) > 0) | (((msa2_status == "UC/LU").sum(axis = 1)) > 0)]
msa1_deliver_status = msa1_status[(((msa1_status == "LU").sum(axis = 1)) > 0) | (((msa1_status == "UC/LU").sum(axis = 1)) > 0)]
msa2_deliver_status = msa2_status[(((msa2_status == "LU").sum(axis = 1)) > 0) | (((msa2_status == "UC/LU").sum(axis = 1)) > 0)]
print(msa1_deliver_occ.shape)
print(msa2_deliver_occ.shape)
print(msa1_deliver_status.shape)
print(msa2_deliver_status.shape)
```

```python
def slicedataset(dataframe, start, end=None):
    if end == None:
        return dataframe[dataframe.columns[start:]]
    else:
        return dataframe[dataframe.columns[start:end+1]]
```

```python
msa1_deliver_occ = slicedataset(msa1_deliver_occ, start_occ, end_occ)
msa2_deliver_occ = slicedataset(msa2_deliver_occ, start_occ, end_occ)
msa1_deliver_status = slicedataset(msa1_deliver_status, start_status, end_status)
msa2_deliver_status = slicedataset(msa2_deliver_status, start_status, end_status)
```

```python
def deliver_date(row):
    for i in range(len(row)):
        if (row[i] == "LU" or row[i] == "UC/LU"):
            return int(i)
def deliver_date_all(data):
    nrow = data.shape[0]
    output = []
    for i in range(nrow):
        output.append(deliver_date(list(data.iloc[i])))
    return output
```

```python
def occ_date(row, start):
    for i in range(start,len(row)):
        if row[i] >= 0.9:
            return int(i)
def occ_date_all(data, startlist):
    nrow = data.shape[0]
    output = []
    for i in range(nrow):
        output.append(occ_date(list(data.iloc[i]), startlist[i]))
    return output
```

```python
import statistics
def leaseup_time(data_occ, data_status):
    data_start = deliver_date_all(data_status)
    data_end = occ_date_all(data_occ, data_start)
    total = 0
    right_censor = 0
    output = []
    for i in range(len(data_start)):
        if (data_end[i] != None):
            output.append(data_end[i] - data_start[i] + 1)
            total += data_end[i] - data_start[i] + 1
        else:
            right_censor += 1
    if (len(data_start) - right_censor == 0):
        print("All observed data are right-censored.")
        return (data_occ.shape[1] - data_start[0] + 2), right_censor, output
    else:
        return total / (len(data_start) - right_censor), right_censor, output
```

```python
msa1_avg, num_censor1, output1 = leaseup_time(msa1_deliver_occ, msa1_deliver_status)
print("Average lease-up time: ", msa1_avg)
print("Sample standard deviation: ", statistics.stdev(output1))
print("Number of right censored data: ", num_censor1)
```

```python
msa2_avg, num_censor2, output2 = leaseup_time(msa2_deliver_occ, msa2_deliver_status)
print("Average lease-up time: ", msa2_avg)
print("Sample standard deviation: ", statistics.stdev(output2))
print("Number of right censored data: ", num_censor2)
```

## Answer:
<p> (1) For market "Austin-Round Rock, TX", there are 270 properties that started to be delivered between Apr 2008 and Sep 2019. Among them, 26 properties haven't reached 90% occupancy up to Sep 2019. In order to receive a more accurate estimate of the average lease-up time, I drop those observations with unobserved lease-up time, and compute the average based on remaining observations (i.e. 244 observations).
<p> For market "Austin-Round Rock, TX", the average lease-up time is 13.7213 months. 
<p> Below is a histogram of the lease-up time computed from 244 observation, which clearly shows that the lease-up time is highly deivated among different properties in the Austin-Round Rock market.

```python
import matplotlib.pyplot as plt
num_bins = 20
plt.hist(output1, num_bins)
plt.xlabel("Lease-up Time")
plt.ylabel("Frequency")
plt.title("Histogram of lease-up time for 'Austin-Round Rock, TX' market")
plt.show()
```

<p> (2) For market "Akron, OH", there were only 9 properties that started to be delivered between Apr 2008 and Sep 2019. Among 9 properties, 1 property had unobserved lease-up time. Therefore, I can only estimate the average lease-up time based on remaining 8 observations. The lack of observations might lead to a highly biased. 
<p> For market "Akron, OH", a highly biased estimate of the average lease-up time is 9.75 months.

## Task 2 Migration Exploration
### 1. Which county pair do you see the most interactions (in terms of number of exemptions)?

```python
# Load all data files in migration folder
# Merge all data files together and store them in migration
import os
files = os.listdir("./raw_data/migration")
files_xls = [f for f in files if f[-3:] == 'xls']
migration = pd.DataFrame()
for f in files_xls:
    temp_data = pd.read_excel("./raw_data/migration/" + f, header = None, skiprows = 8)
    migration = migration.append(temp_data)
```

```python
migration.head()
```

```python
# Find the row with maximum nunmber of exemption
migration[migration[7] == migration[7].max()]
```

<p> However, the county pair found cannot be served as the final answer for this question, as the column 0 and 1 are exactly the same as column 2 and 3. This indicates that this entry cannot be considered as the interaction between different counties. Therefore, I decided to make another restriction to eliminate those situations.

```python
import numpy as np
real_migration = migration[np.invert((migration[0] == migration[2]) & (migration[1] == migration[3]))]
real_migration[real_migration[7] == real_migration[7].max()]
```

<p> Moreover, I found that when the index in column 2 is greater or equal to 57, the corresponding row is the sum values. Therefore, I also need to elininate those rows in my calculation.

```python
real_migration = real_migration[np.invert((real_migration[2] >= 57))]
real_migration[real_migration[7] == real_migration[7].max()]
```

## Answer:
<p> Therefore, based on above calculation, the county pair that I see having the most interactions are county 71 in California and county 37 in California.

### 2. Which county do you see the most in-migration (in terms of number of exemptions)?

```python
in_migration = real_migration.groupby([0, 1]).sum()
in_migration[in_migration[7] == in_migration[7].max()]
```

## Answer:
Based on above calculation, the county that has most in-migration is the the county 37 in California, whose total number of in-migration is $176914$.

### 3. Which county do you see the most out-migration (in terms of number of exemptions)?

```python
out_migration = real_migration.groupby([2, 3]).sum()
out_migration[out_migration[7] == out_migration[7].max()]
```

## Answer:
Based on above calculation, the county that has most out-migration is the the county 37 in California, whose total number of out-migration is $245959$.
